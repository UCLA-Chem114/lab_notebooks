## Sample Project: Smoothing your Data
Moving/rolling data averaging and other approaches to data smoothing

```python
# first init graphics and import libraries
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# make dummy data
x = np.linspace(0,3*np.pi,100)
y = np.cos(x)
z = y + np.random.normal(size=100)*0.3
```

```python
# plot dummy data
plt.style.use('seaborn')

plt.plot(x,y)
plt.plot(x,z,'or');
```

```python
# Use convolve in Numpy
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth

plt.plot(x,z,'or')
plt.plot(x,smooth(z,11));
```

```python
# Savitzky-Golay filter 
from scipy.signal import savgol_filter
yhat = savgol_filter(z, 9, 3) # window size 9, polynomial order 3
plt.plot(x,z,'or')
plt.plot(x,yhat)
```

```python
# Pandas rolling window methods
df = pd.DataFrame(z)
```

```python
df.head()
```

```python
df_mean = df[0].rolling(5).mean()
```

```python
plt.plot(x,z,'or')
plt.plot(x,df_mean)
```
