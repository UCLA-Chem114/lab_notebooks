### 2-D arrays of random numbers
Example of displaying 2-d data as a temperature image or heatmap with a colorbar

```python
# generate 2-d array random numbers
ra = np.random.randn(20,20)  # 20x20 array of random numbers
with plt.style.context('classic'):  # apply classic style to this plot only
    plt.imshow(ra)  # display random numbers as image or 'heatmap'
    plt.colorbar(); # add colorbar
```
