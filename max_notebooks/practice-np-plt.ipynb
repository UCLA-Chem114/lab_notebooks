```python
# import numpy and matplotlib modules
import numpy as np
import matplotlib.pyplot as plt
```

### Basic Numpy and Matplotlib commands

```python
# set display precision for floating numbers and use fixed notation
np.set_printoptions(precision=2,suppress=True);
```

```python
# create and index numpy 1-d arrays
print('** create and index numpy arrays **')
arr1 = np.array([1,3,5,7,9])  # make numpy array from list [1,3,5,7,9]
arr2 = np.array([0,2,4,6,8])  # another from [0,2,4,6,8]
print('arr1: ',arr1)
print('arr2:', arr2)

print('first element arr1[0] : ',arr1[0])
print('last element arr1[-1]: ',arr1[-1])
print('range arr1[:3] ',arr1[:3])
print('range arr1[2:] ',arr1[2:])
print('range arr1[1:3] ',arr1[1:3])

print('\n ** simple operations on arrays **')  # special characters - \n creates a newline
print('sum of arrays: ', arr1+arr2)
print('product of arrays: ', arr1*arr2)
print('arr1 * 3: ', arr1*3)

print('\n ** some array properties **')  # special characters - \n creates a newline
print('arr1 size: ',arr1.size)
print('arr1 dimension: ',arr1.ndim)
print('arr1 shape: ',arr1.shape)
print('arr1 dtype: ',arr1.dtype)

# other ways to create numpy arrays
arr3 = np.linspace(0,9,10)
print(arr3)

# create and print a 2-d array filled with zeros
print('\n ** creating a 3x3 zero and random 2-d arrays **')  # special characters - \n creates a newline
arr4 = np.zeros((4,4))
print(arr4)
arr5 = np.random.rand(4,4)
print(arr5)

# loading a file with data = np.loadtxt also creates a numpy array
# look at the array properties of arr5, try adding/subtracting numbers to arr5
```

```python
# universal functions on arrays
print('cos of arr3: ', np.cos(arr3))
print('exp of arr3: ', np.exp(arr3))
print('sin of 2-d arr5:\n', np.sin(arr5))  # \n character creates a newline

# define an array using list [0, np.pi/6, np.pi/4, np.pi/3, np.pi/2, np.pi] and take sin, cos, tan functions of this array
```

```python
# functions we wrote and tested on numbers also work on numpy arrays
# modularity and abstraction with functions
# we always first define a function and then use it; note mutiline comment in triple quotes

def cube(x):   # like in conditionals and loops : and TABBED IN actions
    '''cube function computes the input number cubed,
       will generate errors on non-number input or empty input'''
    return x**3  # exit the function returning cubed value of input

# 2. function without return, side effect only - print or plot something
def trig(angle):   # like in conditionals and loops : and TABBED IN actions 
    '''trig function tests the trigonometric identity;
       for all numberical inputs this function should return 1., 
       pretty boring, since cos^2(a) + sin^2(a) = 1'''
    return np.cos(angle)**2 + np.sin(angle)**2  # return 1 ?
```

```python
# run the functions defined above
print('cube function works on arrays')
result = cube(arr3)  # run cube on array 3 and store return in the result variable
print( result )   # print result holding the computer cube of array 3

print( cube(arr5)) # compute cube 2-d arr5

print('\ntrig function should return all 1s')
print( trig(0) )              # trig identity works on 0
print( trig(np.pi/2) )        # and on pi/2

print( trig(arr5) )

# write a function that computes x^3 - b*x with inputs x and b
```

```python
# create several arrays to plot
x = np.linspace(0, 3*np.pi,300)  # regularly spaced 300 points between 0 and 3$\pi$
y = np.cos(x)  # cos function applied to x array
z = np.cos(x*x)  # chirp function

print('y and z look the same for x near 0')
print('x :',x[:10])  # print first 10 elements of x
print('y :',y[:10])  # print first 10 elements of y
print('z :',y[:10])  # print first 10 elements of z

# print elements 100-109 for x,y,z ; are y and z still the same?
```

```python
# setup plot style
# plt.style.available[:10]  - some available styles
plt.style.use('seaborn')  # use a particular style
```

```python
# plot y and z vs x
plt.plot(x,y,label='$cos(x)$')  # plot y vs x with math label
plt.plot(x,z,label='$cos(x^2)$') # plot z vs x with math label
plt.legend() # plot legend
plt.xlabel('X (radians)')
plt.ylabel('Amplitude')
plt.title('cos and chirp functions',fontsize=18);  # title with fontsize change
# plt.savefig('plot')
```

```python
# plot a portion the above plot
plt.plot(x,y,x,z)   # display both y vs x and z vs x, can be separate
plt.plot(x,z,'ro')  # display data points as red dot
plt.xlim((3,5))  # limit x axis to 3-5
plt.title('plot detail');

# try other xlim and ylim commands
```
