# The law of large numbers (LLN)

>“…Even the stupidest man knows by some instinct of nature…that the greater the
> number of confirming observations, the surer the conjecture.”
>
>– Jacob Bernoulli
>
>From the [*Ars Conjectandi*](https://en.wikipedia.org/wiki/Ars_Conjectandi) 


In other words, the average of a large number of events should be close to the expected value, $E(X)$, and approaches $E(X)$ as the number of events reaches infinity. 

## Back to our 6-sided dice example

Again, we will simulate a uniform distribution for rolling a 6-sided dice. Each time the dice is rolled, there is an euqal chance of the outcome being a number 1 - 6. 

```python
# define number of times the dice is rolled
num_rolls = 400
num_sides = 6
num_trials = 1000
```

```python
mu = np.sum(np.arange(1,num_sides + 1,1)) / num_sides
mu
```

```python
trial = np.arange(1, num_trials+1, 1)
```

```python
d_avg = []

# simulate n dice rolls for 6-sided dice for t trials 
for t in trial:
    d = np.random.randint(low = 1, high = num_sides + 1, size = t)
    avg = np.mean(d)
    d_avg.append(avg)
```

```python
plt.plot(trial, d_avg, linewidth=2.5, zorder=1, label='average roll #')
plt.axhline(mu, linewidth=2, color='k', linestyle='--', label=f'$\mu$ = {mu}', zorder=2)
plt.xlim(0, num_trials)
plt.ylim(0, num_sides+0.5)
plt.xlabel('trials')
plt.ylabel('average')
plt.title('Law of Large Numbers')
plt.legend();
```
