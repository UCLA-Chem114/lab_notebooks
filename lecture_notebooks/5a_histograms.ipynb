```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'  # use this for hi-dpi displays
sns.set_style('whitegrid')  # set the plotting style
```

# Simulate data

Here, we simulate experimental data for 10,000 measurements of CD diameters with normally distributed random noise

```python
# set random number generator with defined seed value
np.random.seed(0)

# define mean and standard deviation of distribution
mu, sigma, n = 120, 2, 10000

# generate pseudo-random, normally distributed sample data
x = np.random.normal(loc = mu, scale = sigma, size = n)
print(f'mu: {mu: .2f} \n' + f'sigma: {sigma: .2f} \n' + f'x: {x}')
```

# The histogram is a discrete approximation of the probability distribution of a random variable

```python
mean = x.mean()
stdv = x.std()
bins = 20

# plot histogram
plt.hist(x, bins, label=f'n = {n}') 
# plot vertical line at the mean
plt.axvline(mean, color='k', linestyle='dashed', linewidth=1, label = f'$\mu$ = {mean: .2f}') 
# add labels 
plt.title(f'histogram of CD diameter, $\sigma$={stdv: 0.2f}')
plt.xlabel('diameter (mm)')
plt.ylabel('frequency')
plt.legend();
```

# The distribution can be fit using the probability density function (PDF)

In this example, we use a normal (Gaussian) distribution. 
However there are many types of distributions and a PDF can be calculated for each e.g. Cauchy distribution, Rayleigh distribution, and log-normal distribution. 

```python
bins = 200

# plot normalized histogram
plt.hist(x, bins, density=True, label=f'n = {n}')
plt.axvline(mean, color='k', linestyle='dashed', linewidth=1, label = f'$\mu$ = {mean: .2f}')
plt.title(f'histogram of CD diameter, $\sigma$={stdv: 0.2f}')
plt.xlabel('diameter (mm)')
plt.ylabel('probability density')

# add PDF to normalized histogram
bins = np.linspace(110, 130, 200) # create array of bins
bin_centers = 0.5*(bins[1:] + bins[:-1]) # calculate bin centers from array
pdf = stats.norm.pdf(x = bin_centers, loc=mu, scale=sigma) # calculate PDF
plt.plot(bin_centers, pdf, '-', lw =3, label="PDF")
plt.legend();
```

Decreasing the bin size can result in loss of information.

# Increasing the sample size, $N$, improves the histogram approximation 

# For large $N$ and increased number of bins, the histogram converges with the  PDF

```python
# increase N by 10^3
n = 10000000
bins = 200
x = mu + sigma * np.random.randn(n)

# plot normalized histogram
plt.hist(x, bins, density=True, label=f'n = {n}')
plt.axvline(mean, color='k', linestyle='dashed', linewidth=1, label = f'$\mu$ = {mean: .2f}')
plt.title(f'histogram of CD diameter, $\sigma$={stdv: 0.2f}')
plt.xlabel('diameter (mm)')
plt.ylabel('probability density')

# add PDF to normalized histogram
bins = np.linspace(110, 130, 200) # create array of bins
bin_centers = 0.5*(bins[1:] + bins[:-1]) # calculate bin centers from array
pdf = stats.norm.pdf(x = bin_centers, loc=mu, scale=sigma) # calculate PDF
plt.plot(bin_centers, pdf, '--', label="PDF")
plt.legend();
```
